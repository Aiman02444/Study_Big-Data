```python
x = np.linspace(0, 10, 100) #0부터 10까지 100개
y = np.sin(x)
y_ = np.cos(x)

plt.plot(x, y)
plt.plot(x, y_, '-o')
```

```python
x = np.linspace(0, 10, 100) #0부터 10까지 100개
y = np.sin(x)
y_ = np.cos(x)

plt.plot(x, y, label='sin')
plt.plot(x, y_, '-o', label='cos')

plt.legend(loc=4) #loc 는 위치(분면)
```

## scatter

```python
x = np.linspace(0, 10, 20)
y = x ** 2

plt.scatter(x, y, color ='r', alpha=0.5) #산점도
plt.show() #있어도되고 없어도됨
```

## histogram

```python
x = [np.random.randint(1, 7) for i in range(1000000)] #주사위를 10번 던진 수 리스트
x
```

```python
plt.hist(x, bins = 11)
plt.show() #range값, 큰수의 법칙 따름 -> 데이터의 숫자가 많아질수록, 원하는 예측값에 도달할 확률 이 높다
```

## Pie Chart

```python
labels = ['one', 'two', 'three']
size = [100, 20, 10] #100/130, 20/130, 10/130 을 백분율로 표시

plt.pie(size, labels = labels)
plt.show()
```

```python
labels = ['one', 'two', 'three']
size = [100, 20, 10]

plt.pie(size, labels = labels, shadow = True) #밑에 그림자가 살짝 생김
plt.show()
```

```python
labels = ['one', 'two', 'three']
size = [50, 20, 30]

plt.pie(size, labels = labels, shadow = True, autopct = '%1.2f%%')
plt.show()
```

## Bar Chart

```python
plt.bar(['one', 'two', 'three'], [10, 20, 30])
plt.show()
```

```python
plt.barh(['one', 'two', 'three'], [10, 20, 30])
plt.show()
```
