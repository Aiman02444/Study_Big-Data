```python
sample2 = sample[::10, ::10]
```

```python
plt.hist(sample2.ravel(), 256, [0, 256]) #색의 분포
plt.show()
```

```python
sample2_ = np.where(sample2 < 50, sample2, 0)
plt.imshow(sample2_)
```
